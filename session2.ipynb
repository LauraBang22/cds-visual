### Defining functions in Python

```python
def hello():
    #alt hvad der kommer under funktioner er hvad den gør
    print("Hello, world!")
```

```python
#Se funktionen
hello()
```

```python
#Få funktionen til at sige hej til et navn
def hello(name):
    print("Hello, " + name + "!")
```

```python
hello("Laura")
```

```python
def hello(name= "world"):
    print("Hello, " + name + "!")

#Hvis man ikke putter noget i funtionen printer den world, frem for en error besked
```

```python
hello()
```

```python
### Good Python practice
```

```python
#skriv i koden hvad det er den gør
def hello(name= "world"):
    """
    This function takes a name and prints message to screen
    """
    print("Hello, " + name + "!")
```
